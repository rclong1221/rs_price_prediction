# Model Tuning SVM

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import svm

warnings.filterwarnings('ignore')
%matplotlib inline

SEED=42
```

```python
ci_i5_ap = pd.read_csv('./data/ci_i5_ap.csv')
```

```python
ci_i5_ap_train, ci_i5_ap_test, ci_i5_ap_train_Y, ci_i5_ap_test_Y = train_test_split(ci_i5_ap, ci_i5_ap['price'], test_size=0, random_state=42)
ci_i5_ap_train.drop('price', axis=1, inplace=True)
ci_i5_ap_train.drop('date', axis=1, inplace=True)
ci_i5_ap_train.drop('id', axis=1, inplace=True)

scaler = StandardScaler()
ci_i5_ap_train = scaler.fit_transform(ci_i5_ap_train)
```

```python
# ci_i5_ap_train.dtypes
```

```python
params = {
    'objective': ['reg:linear'],
    'seed': [SEED],

    # 1
    'degree': range(0, 3, 1),
    
    # 2
    'C': [0.001, 0.003, 0.01, 0.03, 0.1, 0.3, 1],
    
    # 3
    'learning_rate': [0.01]
}
```

### degree

```python
scores = dict()

for degree in params['degree']:
    if 'degree' not in scores:
        scores['degree'] = dict()
    svr = svm.SVR(degree=degree)
    cv = cross_val_score(svr, ci_i5_ap_train, ci_i5_ap_train_Y, cv=10, scoring='neg_median_absolute_error')
    scores['degree'][degree] = cv.mean()

print(scores)
```

### C

```python
backup1 = scores.copy()

for C in params['C']:
    if 'C' not in scores:
        scores['C'] = dict()
    svr = svm.SVR(degree=1, C=C)
    cv = cross_val_score(svr, ci_i5_ap_train, ci_i5_ap_train_Y, cv=10, scoring='neg_median_absolute_error')
    scores['C'][C] = cv.mean()

print(scores)
```

### $-MAE$ respective to degree

```python
# convert to pandas dataframe
md = {'Degree': list(scores['degree'].keys()), '-MAE': list(scores['degree'].values())}
md_df = pd.DataFrame(md)

# plot using lineplot
sns.set(style='darkgrid')
sns.lineplot(x='Degree', y='-MAE', data=md_df).set_title("-MAE According to Degree")
```

### $-MAE$ respective to C

```python
# convert to pandas dataframe
md = {'C': list(scores['C'].keys()), '-MAE': list(scores['C'].values())}
md_df = pd.DataFrame(md)

# plot using lineplot
sns.set(style='darkgrid')
sns.lineplot(x='C', y='-MAE', data=md_df).set_title("-MAE According to C")
```
